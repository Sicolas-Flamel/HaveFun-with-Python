```python
# 读数据
data = ''
with open(r'C:\Users\ASUS\Desktop\zijie.txt', 'r',encoding='utf-8') as f:   
    for line in f.readlines():
        data += line
```

```python
data.split('\n')  
```

```python
import re
```

```python
p1 = re.compile('([a-zA-Z]{3,10})')
```

```python
skill = re.findall(p1,data)
```

```python
# 消除大小写的影响，避免 把  Excel 和 excel 看作是两种技能
for i in range(len(skill)):
    skill[i] = skill[i].upper()
```

```python
skill
```

```python
from collections import Counter
```

```python
skill_nums = Counter(skill)
```

```python
skill_nums = dict(skill_nums)
```

```python
del skill_nums['APP']  # 删掉一些明知道不是技能的词
```

```python
from pandas import DataFrame

skill_nums = DataFrame(skill_nums, index=range(1))
skill_nums = skill_nums.T
```

```python
skill_nums = skill_nums.sort_values(by = 0, ascending =False)
```

```python
fig = plt.figure(dpi=80, figsize=(12, 8))
plt.title("技能统计")
plt.plot(skill_nums[skill_nums[0]>5].index, skill_nums[skill_nums[0]>5][0],'-o', color = "darkslateblue", label="出现次数")
plt.legend()
fig.autofmt_xdate(rotation=30)
```

```python
fig = plt.figure(dpi=80, figsize=(12, 8))
plt.title("技能统计")
plt.plot(skill_nums.index, skill_nums[0],'-o', color = "y", label="出现次数")
plt.legend()
fig.autofmt_xdate(rotation=90)
```
